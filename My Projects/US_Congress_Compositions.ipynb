```python
import csv
import numpy as np
import pandas as pd
```

```python
#open file
congress_comp = pd.read_csv('US_Congress_Compositions.csv', delimiter=',')

#convert floats to ints
pd.options.display.float_format = '{:.0f}'.format
```

```python
#Compare Senate and House differences
congress_comp['Senate_Diff'] = congress_comp['Senate_D'] - congress_comp['Senate_R']
congress_comp['House_Diff'] = congress_comp['House_D'] - congress_comp['House_R']
```

```python
#Remove third party columns and rows with zeroes in D# or R# columns
congress_comp_diff = congress_comp.loc[:, ['Year', 'President', 'Senate_Diff', 'House_Diff']]
congress_comp_diff = congress_comp_diff.dropna()

congress_comp_diff
```

```python
congress_comp_dem_senate = congress_comp_diff.loc[:, ['Year', 'President', 'Senate_Diff']]
congress_comp_dem_senate = congress_comp_dem_senate.where(congress_comp_diff['Senate_Diff'] > 0)
congress_comp_dem_senate = congress_comp_dem_senate.dropna()
congress_comp_dem_senate
```

```python
congress_comp_rep_senate = congress_comp_diff.loc[:, ['Year', 'President', 'Senate_Diff']]
congress_comp_rep_senate = congress_comp_rep_senate.where(congress_comp_diff['Senate_Diff'] <= 0)
congress_comp_rep_senate = congress_comp_rep_senate.dropna()
congress_comp_rep_senate
```

```python
congress_comp_dem_house = congress_comp_diff.loc[:, ['Year', 'President', 'House_Diff']]
congress_comp_dem_house = congress_comp_dem_house.where(congress_comp_diff['House_Diff'] > 0)
congress_comp_dem_house = congress_comp_dem_house.dropna()
congress_comp_dem_house
```

```python
congress_comp_rep_house = congress_comp_diff.loc[:, ['Year', 'President', 'House_Diff']]
congress_comp_rep_house = congress_comp_rep_house.where(congress_comp_diff['House_Diff'] < 0)
congress_comp_rep_house = congress_comp_rep_house.dropna()
congress_comp_rep_house
```
